```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
def prepare(x_train):
    x_train['intercept']=1
    scaler=StandardScaler()
    scaler.fit(x_train)
    x_train=scaler.transform(x_train)
    x_train[:,-1]=1
    return x_train
```

```python
def slope(x,y,j,m):
    c=0
    n=len(x)
    for i in range(n):
        pred=0
        for k in range(len(x[0])):
            pred=pred+m[k]*x[i,k]
        c=c+(y[i]-pred)*2*-1*x[i][j]
    c=c/n
    return c
```

```python
def predict(x,m):
    yp=[]
    n=len(x)
    for i in range(n):
            r=0
            for k in range(len(m)):
                r=r+m[k]*x[i,k]
            yp.append(r)
    return np.array(yp)
```
